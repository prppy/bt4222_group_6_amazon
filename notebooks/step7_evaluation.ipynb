# step5_4_Hybrid_NCF_Model Results
The Hybrid NCF Model combines two parallel Neural Collaborative Filtering models: one using custom embeddings derived from engineered user and product features, and the other using randomly initialised embeddings. Their outputs are merged through a fully connected layer, allowing the model to learn a weighted combination of both representations. This architecture is designed to capture both structured patterns and latent interactions for improved recommendation performance.

```python
# step5_4_Hybrid_NCF_Model Results

base_dir = '/content/drive/MyDrive/bt4222_group_6/bt4222_group_6_amazon/Model Results/NCF Hybrid Embedding'
cluster_ids = sorted([int(d.split('_')[1]) for d in os.listdir(base_dir) if d.startswith('Cluster_')])

# Collect results
val_results_all = []
test_results_all = []

for cid in cluster_ids:
    cluster_dir = os.path.join(base_dir, f'Cluster_{cid}')

    # Load val results
    val_path = os.path.join(cluster_dir, 'val_results_by_cluster.csv')
    if os.path.exists(val_path):
        df_val = pd.read_csv(val_path)
        val_results_all.append(df_val)

    # Load test results
    test_path = os.path.join(cluster_dir, 'test_results_by_cluster.csv')
    if os.path.exists(test_path):
        df_test = pd.read_csv(test_path)
        test_results_all.append(df_test)

# Combine into two dataframes
val_df = pd.concat(val_results_all, ignore_index=True)
test_df = pd.concat(test_results_all, ignore_index=True)

# Print Results
print("========step5_4_Hybrid_NCF_Model Validation Results by Cluster:========")
display(val_df)

print("========step5_4_Hybrid_NCF_Model Test Results by Cluster:========")
display(test_df)

print("Average Validation Metrics:")
print(val_df.mean(numeric_only=True))

print("\nAverage Test Metrics:")
print(test_df.mean(numeric_only=True))
```

#step6_1_1- LSTM Model Results

The LSTM model predicts the likelihood of a user purchasing a candidate product based on their past interactions. Each item in the user's purchase history is embedded into a dense vector and passed through a two-layer LSTM to capture sequential patterns in user behavior. The final hidden state of the LSTM is concatenated with the embedding of the candidate product, forming a combined representation. This is passed through a fully connected layer followed by a sigmoid activation to output the probability of purchase.

```python
#step6_1_1- LSTM Model
LSTM_val_results = pd.read_csv(os.path.join(results_dir,"LSTM/val_evaluation_results.csv"))
print(LSTM_val_results.shape)
print("==============step6_1_1- LSTM Model Validation Results===============")
display(LSTM_val_results.head())

LSTM_test_results = pd.read_csv(os.path.join(results_dir,"LSTM/test_evaluation_results.csv"))
print(LSTM_test_results.shape)
print("==============step6_1_1- LSTM Model Testing Results===============")
display(LSTM_test_results.head())
```

#step6_1_2- LSTM Model: Custom Embedding

The LSTM Model with custom embedding enhances purchase prediction by combining sequential modeling with user- and product-specific feature embeddings. It processes the user's purchase history through a multi-layer LSTM to capture temporal behavior patterns. The final hidden state of the LSTM is concatenated with the embedding of the candidate product, as well as dense vectors representing custom user and product features. The combined representation is passed through a fully connected layer with a sigmoid activation to produce the final purchase probability.
